```python
#data preprocessing: handling NA values
#Linear multivarite regression
import pandas as pd
import numpy as np
from sklearn import linear_model
df = pd.read_csv("homeprice01.csv")
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2500,4,5]])
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
df = pd.read_csv("Salary prediction.csv")
df
```

```python
import math
median_test_score_value = math.floor(df['test_score(out of 10)'].median())
median_test_score_value
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_test_score_value)
df
```

```python
df['Employe Experience'] = df['Employe Experience'].fillna("zero")
df
```

```python
from word2number import w2n
df['Employe Experience'] = df['Employe Experience'].apply(w2n.word_to_num)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Employe Experience','test_score(out of 10)','interview_score(out of 10)']],df.Salary)
```

```python
reg.coef_
```

```python
reg.intercept_
```
