```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==1].head()
```

```python
df['flower name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
df2.head()
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='blue',marker='*')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='maroon',marker='>')
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='black',marker='*')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='green',marker='>')
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df.target
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
model = SVC(kernel="linear")
```

```python
model.fit(x_train, y_train) ## in this model is going to automatically take kernel rbf how it is?
```

```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.target
```

```python
dir(digits)
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
digits.target_names
```

```python
df[df.target==1].head()
```

```python
df['digits'] = df.target.apply(lambda x: digits.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
df2.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3)
```

```python
from sklearn.svm import SVC
rbf_model = SVC(kernel='rbf')
```

```python
len(X_test)
```

```python
rbf_model.fit(X_train, y_train)
```

```python
rbf_model.score(X_test,y_test)
```

```python
linear_model = SVC(kernel='linear')
```

```python
linear_model.fit(X_train,y_train)
```

```python
linear_model.score(X_test,y_test)
```

```python
## adding kNN algorthim
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=10)
KNN.fit(X_train, y_train)
```

```python
KNN.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = KNN.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
